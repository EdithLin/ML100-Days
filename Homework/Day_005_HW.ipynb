```python
#設定資料讀取路徑 並讀取資料
dir_path = '/home/edith/文件/100days-data'
data_app = os.path.join(dir_path, 'application_train.csv')
data_app2 = pd.read_csv(data_app)
data_app2.head()
```

```python
#先觀察資料外觀
data_app3 = pd.DataFrame(data_app2)
data_app3.shape #看資料總共幾欄幾行
data_app3.describe() #可以看到欄位各項數值（已計算完的數值）
```

#### 計算集中趨勢

```python
#平均值，計算AMT_INCOME_TOTAL的平均值
data_app3['AMT_INCOME_TOTAL'].mean()
```

```python
#計算AMT_INCOME_TOTAL的變異數
data_app3['AMT_INCOME_TOTAL'].var()
```

```python
#計算AMT_INCOME_TOTAL的中位數
data_app3['AMT_INCOME_TOTAL'].median()
```

```python
#眾計算AMT_INCOME_TOTAL的眾數
data_app3['AMT_INCOME_TOTAL'].mode()
```
